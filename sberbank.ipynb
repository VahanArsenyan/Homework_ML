```python
data = pd.concat([test, train], sort=False)
```

```python
data_1 = data.iloc[:,1:13]
data_1['price'] = data['price_doc']
data_1 = data_1.dropna(how='any')
```

```python
print(data.shape)
print(data_1.shape)
```

```python
data_1["timestamp"] = pd.to_datetime(data_1["timestamp"], infer_datetime_format=True)
```

```python
data_1.head()
```

```python
data_2 = data_1.loc[(data_1['build_year'] < 2017) & (data_1['build_year'] > 1930)]
```

```python
data_2.shape
```

```python
data_2.describe()
```

```python
data_2.dtypes
```

```python
data_2['timestamp'] = (data_2['timestamp'].dt.year * 12 + data_2['timestamp'].dt.month)*4
```

```python
data_2.head()
```

```python
repl_pt = {v:k for k,v in enumerate(data_2['product_type'].unique())}
print(repl_pt)
```

```python
data_2['product_type'] = data_2['product_type'].replace(repl_pt)
```

```python
data_2.head(1)
```

```python
repl_pt = {v:k for k,v in enumerate(data_2['sub_area'].unique())}
print(repl_pt)
```

```python
data_2['sub_area'] = data_2['sub_area'].replace(repl_pt)
```

```python
data_2.head(1)
```

```python
data_2.shape
```

```python
import xgboost as xgb
```

```python
xg_cl = xgb.XGBClassifier(objective='reg:linear', n_estimators=200, seed=123, max_depth=3, nthread=8)
```

```python
xg_cl.fit(data_2.iloc[:10000,:-1], data_2.iloc[:10000,-1])
```

```python
res = xg_cl.predict(data_2.iloc[10000:,:-1])
```

```python
np.sqrt(np.mean((np.log(res) - np.log(data_2.iloc[10000:,-1]))**2))
```
